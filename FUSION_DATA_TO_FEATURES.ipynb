```python
import pickle as pkl
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
with open('/content/vectorizer_text.pkl','rb') as f:
  vectorizer = pkl.load(f)
```

```python
data=pd.read_pickle('/content/fusion_data.pkl')
```

```python
data.head()
```

```python
text_vectors = vectorizer.transform(data["text"]).toarray()
```

```python
data['text']=list(text_vectors)
```

```python
data
```

```python
data['text'][0].shape
```

```python
import tensorflow as tf

text_model = tf.keras.models.load_model("/content/text_model.keras")

emotion_model = tf.keras.models.load_model("/content/ravdees-2.keras")

sound_model = tf.keras.models.load_model("/content/background_model.keras")
```

```python
text_model.summary()
#feature layer is -2
```

```python
emotion_model.summary()
#feature layer is -3
```

```python
sound_model.summary()
#feature layer is -3
```

```python
dummy = np.zeros((1, len(data["text"][0])))
text_model.predict(dummy)
```

```python
text_feature_model = tf.keras.Model(
    inputs=text_model.inputs,
    outputs=text_model.layers[-2].output
)
```

```python
dummy = np.zeros((1,128,200,1))
emotion_model.predict(dummy)

emotion_feature_model = tf.keras.Model(
    inputs=emotion_model.inputs,
    outputs=emotion_model.layers[-3].output
)
```

```python
dummy = np.zeros((1,128,174,1))
sound_model.predict(dummy)

sound_feature_model = tf.keras.Model(
    inputs=sound_model.inputs,
    outputs=sound_model.layers[-3].output
)
```

```python
sound_feature_model.summary()
```

```python
emotion_feature_model.summary()
```

```python
text_feature_model.summary()
```

GENERATING TEXT FEATURES

```python
text_vectors = np.stack(data["text"].values)
```

```python
text_features = text_feature_model.predict(text_vectors, verbose=1)
```

```python
feature_data=pd.DataFrame()
```

```python
feature_data['text']=list(text_features)
```

EMOTION FEATURES

```python
emotion_mfcc=[]
```

```python
for mfcc in tqdm(data['emotion_features']):
  mfcc=(np.array(mfcc)).reshape(1,128,200,1)
  features=emotion_feature_model.predict(mfcc,verbose=0)
  emotion_mfcc.append(features[0])
```

```python
emotion_mfcc[0].shape
```

```python
feature_data['emotion']=emotion_mfcc
```

BACKGROUND SOUNF FEATURES


```python
sound_feature=[]
```

```python
data['bg_features'][0].shape
```

```python
for mel in tqdm(data['bg_features']):
  mel = np.array(mel).reshape(1,128,174,1)
  features=sound_feature_model.predict(mel,verbose=0)
  sound_feature.append(features[0])
```

```python
np.array(data['bg_features'][0]).shape
```

```python
feature_data['sound']=sound_feature
```

```python
print(len(feature_data['text'][0]))
print(len(feature_data['emotion'][0]))
print(len(feature_data['sound'][0]))
```

```python
feature_data['label'] = data['label']
feature_data['urgency'] = data['urgency']
```
